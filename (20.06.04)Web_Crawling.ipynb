## 기본사용법
- urllib.parse.urlencode()
- urllib.request.urlopen() : urllib.request.Request
- response.getcode() : 200 / 404
- response.read() : html읽기

```python
# import urllib.request
import urllib.parse
import ssl
```

주소를 던져서 열기 (html이 오고 그안에 모든 소스가옴)

```python
#주소
url = 'https://book.naver.com/search/search.nhn'
#파라미터-(query string)
values = {'sm':'sta_hty.book','sug':'','where':'nexearch','query':'파이썬'}
params = urllib.parse.urlencode(values)
#원래 주소처럼 만들기
url = url + '?' + params     
print(url)

context = ssl._create_unverified_context()
response = urllib.request.urlopen(url,context=context)
status_code = response.getcode()
data = response.read()
text = data.decode('utf-8')
print(text)
```

## header 값 추가하기

## requests
- requests.get(): 요청 , headers
- requests.status_code
- requests.text
- get 요청 : params
- post 요청 : data

urllib와 requests의 차이확인
requests 가 더 간편함

```python
import requests
```

```python
# get : params
url = 'https://book.naver.com/search/search.nhn'
#파라미터-(엔코딩 알아서 해버림)
params = {'sm':'sta_hty.book','sug':'','where':'nexearch','query':'파이썬'}
#get()
resp = requests.get(url,params=params)
status_code = resp.status_code
data = resp.text
print(data)
```

```python
#post : data
url = 'https://book.naver.com/search/search.nhn'
#파라미터-(엔코딩 알아서 해버림)
params = {'sm':'sta_hty.book','sug':'','where':'nexearch','query':'파이썬'}
#get()
resp = requests.get(url,data=params)
status_code = resp.status_code
data = resp.text
print(data)
```

```python
#header추가
url = 'https://book.naver.com/search/search.nhn'
#파라미터-(엔코딩 알아서 해버림)
params = {'sm':'sta_hty.book','sug':'','where':'nexearch','query':'파이썬'}
#모바일 header
header_info = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1'}

#get(url,GET,HEADER)
resp = requests.get(url, headers=header_info, params=params)
# resp = requests.get(url, params=params)

status_code = resp.status_code

if status_code == 200:
    resp_url = resp.url
    data = resp.text
    print(resp_url)
#     print(data)    
```

# find() 요소찾기

### 하나만 검색 : find('태그명')

```python
find_h3 = soup.find('h3')
print(find_h3)
print(find_h3.text)
```

### 모두 검색 : find_all('태그명')

```python
find_all_meta = soup.find_all('meta',limit=3)   
# print(find_all_meta)  #list로 옴
for item in find_all_meta:
    print(item)
```

### class로 가져오기
- class_속성사용

```python
item = soup.find(class_="thumb_type thumb_type2")
#a태그 값 가져오기
print(item.find('a')) 
#a태그 안 속성중 href값 가져오기  (책정보주소)
print(item.find('a').attrs['href'])  
print(item.a.attrs['href'])  
```

책 상세정보 링크 담기

```python
find_all_class = soup.find_all(class_="thumb_type thumb_type2")
# len(find_all_class)    #20권
book_detail_url = []
for item in find_all_class:
    book_detail = item.a['href']
    book_detail_url.append(book_detail)
print(book_detail_url)
```

```python
for item in find_all_class:
    book_address = [item.a.attrs['href'] for item in find_all_class]
print(book_address)
```

 책제목 담아보기

```python
find_all_class = soup.find_all(class_="thumb_type thumb_type2")
book_name_list = []
for item in find_all_class:
    book_name = item.a.img['alt']
    book_name_list.append(book_name)
print(book_name_list)
```

BOOK ID 담아보기

```python
find_all_class = soup.find_all(class_="thumb_type thumb_type2")
book_id_list = []
for item in find_all_class:
    book_id = item.a['href'].split('?')[1].split('=')[1]
    book_id_list.append(book_id)
print(book_id_list)
```
