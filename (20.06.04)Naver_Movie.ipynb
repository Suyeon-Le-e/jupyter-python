# 네이버 영화 순위 검색

```python
import requests
from bs4 import BeautifulSoup
```

### 조회순

```python
# 주소작업
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
params = {'sel':'cnt','date':'20200603'}
```

```python
# get 요청
response = requests.get(url,params=params)
status_code = response.status_code
print(status_code)
if status_code ==200:
    text = response.text
```

```python
# str -> BeautifulSoup
soup = BeautifulSoup(text)
```

```python
# 영화 조회수 목록 1개 가져오기
movie_view_all = soup.find('tbody')
movie_view_all
```

```python
movie_tr_all = movie_view_all.select('tr')

movie_tr_all[1]
```

```python
# 제목 뽑기 
title = movie_tr_all[1].find(class_='title').a['title']
title

rank = []
for i in range(1,len(movie_tr_all)-1):
    title = movie_tr_all[i].find(class_='title').a['title']
    if movie_tr_all[i] != none:
        rank.append(title)
rank
```
