```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
# 주소작업
news_id = '217101'
url = 'https://www.wikitree.co.kr/articles/{}'.format(news_id)
resp = requests.get(url)
soup = BeautifulSoup(resp.text)
```

```python
soup
```

```python
names = soup.select('strong')
```

```python
name_list = []
for item in names :
    text_item = item.text
    if text_item == '' :
        continue
    elif '회' in text_item:
        name = text_item.split('회 ')[1]
        name_list.append(name)
```

```python
name_list
```

```python
contents = soup.select('p')
del contents[:4]
del contents[-11:]
del contents[3]
clean_content = []
for item in contents:
    text_item = item.text.replace('\xa0','')
    if text_item == '':
        continue
    else :
        clean_content.append(text_item)
clean_content
```

```python
address_list = []
tel_list = []
for i in range(len(clean_content)):
    if i%3 == 1:
        tel = clean_content[i]
        tel_list.append(tel)
    if i%3 == 2:
        address = clean_content[i]
        address_list.append(address)
        
        
```

```python
print(tel_list)
print(address_list)
```

```python
restaurant_list = []
for i in range(len(address_list)):
    restaurant = (name_list[i],tel_list[i],address_list[i])
    restaurant_list.append(restaurant)
restaurant_list
```
